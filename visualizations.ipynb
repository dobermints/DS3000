```python
import pandas as pd
from geopy.geocoders import Nominatim
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
data = pd.read_csv('berdo/berdo-2016.csv')

# Create a geocoder instance
geolocator = Nominatim(user_agent='my_geocoder')
```

```python
# Function to get coordinates (latitude and longitude) from address
def get_coordinates(address):
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
```

```python
# Apply the function to convert addresses to coordinates and add them as new columns
data['Latitude'], data['Longitude'] = zip(*data['Address'].apply(get_coordinates))

# Filter out rows with missing coordinates
data = data.dropna(subset=['Latitude', 'Longitude'])
```

```python
boston_map = gpd.read_file('path_to_boston_map.geojson')
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data.Longitude, data.Latitude))
fig, ax = plt.subplots(figsize=(10, 8))
boston_map.plot(ax=ax, color='lightgray')
gdf.plot(ax=ax, color='red', markersize=10)
plt.title('Geospatial Plot of Addresses in Boston')
plt.show()
```
