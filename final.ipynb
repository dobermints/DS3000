```python
import pandas as pd
import geopandas as gpd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import seaborn as sns
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
# connect
con = sqlite3.connect('berdo_better.db 2')
con
```

```python
query_1 = """select "GHG Intensity (kgCO2/sf)", ZIP
from berdo_better_full
where length(ZIP) = 5
  and "GHG Intensity (kgCO2/sf)" > 0
group by ZIP
order by "GHG Intensity (kgCO2/sf)" desc;"""

data_1 = pd.read_sql_query(query_1, con)
d = data_1.drop([0, 1, 2, 3, 26])
```

```python
d = d.reset_index()
```

```python
d.drop(['index'], inplace = True, axis = 1)
```

```python
d
```

```python
df = pd.DataFrame(d)

# Create a map centered on Boston
boston_map = folium.Map(location=[42.3601, -71.0589], zoom_start=12)

# Initialize geocoder with rate limiting
geolocator = Nominatim(user_agent="my_app")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Batch geocode zip codes
locations = df['ZIP'].apply(lambda x: geocode(x + ', Boston, MA')).dropna()
```

```python
# Normalize GHG intensity values for marker size
min_intensity = df['GHG Intensity (kgCO2/sf)'].min()
max_intensity = df['GHG Intensity (kgCO2/sf)'].max()
df['Marker Size'] = df['GHG Intensity (kgCO2/sf)'].apply(lambda x: (x - min_intensity) / (max_intensity - min_intensity))  


# Add markers to the map with varying sizes
for location, row in zip(locations, df.iterrows()):
    folium.CircleMarker(
        location=[location.latitude, location.longitude],
        radius=row[1]['Marker Size'] * 10,  # Adjust the size scaling factor as desired
        popup=row[1]['ZIP'] + ' - GHG Intensity: ' + str(row[1]['GHG Intensity (kgCO2/sf)']),
        fill=True,
        fill_color='red',
        color='red',
        fill_opacity=0.7
    ).add_to(boston_map)

# Display the map
boston_map
```
